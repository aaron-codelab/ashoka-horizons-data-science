```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the csv files as a dataframe
df_am = pd.read_csv('all-menu.csv')
```

```python
#Showing top 5 entries
df_am.head(5)
```

```python
#Showing bottom 5 entries
df_am.tail(5)
```

```python
#Checking no. of columns and rows
df_am.shape
```

```python
#Finding the datatype, unique and number of unique entries in each column
for i in range(0, len(df_am.columns), 1):
    print('Datatype of ', df_am.columns[i], ':', df_am.dtypes[i], '\n')
    print('Unique entries in ', df_am.columns[i], ':', df_am[df_am.columns[i]].unique(), '\n')
    print('No. of unique entries in ', df_am.columns[i], ':', df_am[df_am.columns[i]].nunique(), '\n')
```

```python
df_am.isnull().sum()
```

```python
#Finding out null values in each column and creating a dataset without it
df_am_new = df_am
for i in range(0, len(df_am_new.columns), 1):
    for j in range(0, len(df_am_new[df_am_new[df_am_new.columns[i]].isnull()].index), 1):
        df_am_new = df_am_new.drop(df_am_new[df_am_new[df_am_new.columns[i]].isnull()].index[j])
print(df_am_new.shape)
```

```python
#Renaming columns
df_am_new = df_am_new.rename(columns={' Total Fat (g)': 'Total Fat (g)'})
df_am_new = df_am_new.rename(columns={'Trans Fat (g) ': 'Trans Fat (g)'})
df_am_new = df_am_new.rename(columns={' Sodium (mg)': 'Sodium (mg)'})
```

```python
df_am_new.head(5)
```

```python
#Dropping columns
df_am_new = df_am_new.drop(columns=['Saturated Fat (g)'])
df_am_new = df_am_new.drop(columns=[' Dietary Fibre (g)'])
```

```python
df_am_new.head(5)
```

```python
#Finding the Minumum, Maximum, Mean, Standard Deviation, Variance and Sum of that each column with numerical values
for i in range(0, len(df_am.columns), 1):
    if df_am[df_am.columns[i]].dtype == 'int64':
        print('Minimum of column ', i+1, ': ', df_am[df_am.columns[i]].min())
        print('Maximum of column ', i+1, ': ', df_am[df_am.columns[i]].max())
        print('Mean of column ', i+1, ': ', df_am[df_am.columns[i]].mean())
        print('Standard deviation of column ', i+1, ': ', df_am[df_am.columns[i]].std())
        print('Variance of column ', i+1, ': ', df_am[df_am.columns[i]].var())
        print('Sum of column ', i+1, ': ', df_am[df_am.columns[i]].sum())
        print('\n')
```

```python
#Randomly sampling 50% of new data 
df_am_split1 = df_am_new.sample(frac = 0.5)
df_am_split2 = df_am_new
for i in df_am_split1.index:
    df_am_split2 = df_am_split2.drop(i)
```
