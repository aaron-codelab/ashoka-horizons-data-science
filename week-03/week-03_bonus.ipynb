```python
#Importing the necessary packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset

df = pd.read_csv('Quantity of each ingredient.csv')
```

```python
#Showing top 5 entries

df.head(5)
```

```python
#Showing bottom 5 entries

df.tail(5)
```

```python
#Checking the datatype of each column

df.dtypes
```

```python
#Finding missing values

df.isna().sum()
```

```python
df.head(1)
```

```python
#String manipulation and data conversion


for col in df.columns:
    if col in 'INGREDIENTS':
        continue
    df[col] = df[col].str.replace('l', '')
    df[col] = df[col].str.replace('m', '')
    df[col] = df[col].str.replace('g', '')
    df[col] = df[col].astype(float)
```

```python
df.dtypes
```

```python
#Line chart

plt.subplot(1,2,1)
plt.plot(df['INGREDIENTS'], df['QUANTITY PER DAY'], marker='o', color='red', markeredgewidth='5.0', markeredgecolor='blue', markerfacecolor='blue')
plt.title('INGREDIENTS vs QUANTITY PER DAY')
plt.xlabel('INGREDIENTS')
plt.ylabel('QUANTITY PER DAY')

plt.subplot(1,2,2)
plt.plot(df['INGREDIENTS'], df['QUANTITY PER DAY'], marker='o', color='red', markeredgewidth='5.0', markeredgecolor='blue', markerfacecolor='blue')
plt.title('INGREDIENTS vs QUANTITY PER DAY')
plt.xlabel('INGREDIENTS')
plt.ylabel('QUANTITY PER DAY')
```

```python
#Bar chart

plt.subplot(1, 2, 1)
plt.bar(x=df['QUANTITY PER DAY'].unique(), height=df['QUANTITY PER DAY'].value_counts().to_list(), color='#0f5a3e')
plt.xlabel('QUANTITY PER DAY')
plt.ylabel('Value')
plt.title('Bar chart')

plt.subplot(1,2,2)
plt.plot(df['INGREDIENTS'], df['QUANTITY PER DAY'], marker='o', color='red', markeredgewidth='5.0', markeredgecolor='blue', markerfacecolor='blue')
plt.title('INGREDIENTS vs QUANTITY PER DAY')
plt.xlabel('INGREDIENTS')
plt.ylabel('QUANTITY PER DAY')
```

```python
#Area Chart

plt.subplot(1,2,1)
plt.plot(df['INGREDIENTS'], df['QUANTITY PER DAY'], marker='o', color='red', markeredgewidth='5.0', markeredgecolor='blue', markerfacecolor='blue')
plt.title('INGREDIENTS vs QUANTITY PER DAY')
plt.xlabel('INGREDIENTS')
plt.ylabel('QUANTITY PER DAY')
plt.fill_between(df['INGREDIENTS'], df['QUANTITY PER DAY'], color='#ef5b2c')

plt.subplot(1,2,2)
plt.plot(df['INGREDIENTS'], df['QUANTITY PER DAY'], marker='o', color='red', markeredgewidth='5.0', markeredgecolor='blue', markerfacecolor='blue')
plt.title('INGREDIENTS vs QUANTITY PER DAY')
plt.xlabel('INGREDIENTS')
plt.ylabel('QUANTITY PER DAY')
```
